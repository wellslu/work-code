# 類別次數字典1(主頻道分類)

```python
cato_all=[]
for i in tqdm_notebook(cat1['masterChannelTitle']):
    cato_all.append(i)

cato_count=[]
cato_alls=list(set(cato_all))
for i in tqdm_notebook(cato_alls):
    cato_count.append(cato_all.count(i))

c_count1=dict(zip(cato_alls,cato_count))
```

```python
cato_all=[]
for i in tqdm_notebook(cat1['subChannelTitle']):
    cato_all.append(i)
cato_count=[]
cato_alls=list(set(cato_all))
for i in tqdm_notebook(cato_alls):
    cato_count.append(cato_all.count(i))

sub_count1=dict(zip(cato_alls,cato_count))
```

# 主頻道分析

```python
def abc1(aa):
    caty=[]
    for i in tqdm_notebook(aa['pid']):
        try:
            caty.append(cat1[cat1['id'].isin([str(i)])].reset_index(drop=True)['masterChannelTitle'][0])
        except:
            1
    cat_count=[]
    caty_all=list(set(caty))
    for i in tqdm_notebook(caty_all):
        cat_count.append(caty.count(i))
    county=pd.DataFrame({'word':caty_all,
                        'count':cat_count})
    county=county.sort_values(by=['count'],ascending=False).reset_index(drop=True)
    aaa=[]
    for i in tqdm_notebook(range(len(county['count']))):
        aaa.append(county['count'][i]/c_count1[county['word'][i]])
    county['平均點擊']=aaa
    return county
```
