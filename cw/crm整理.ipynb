```python
import pandas as pd
head=['CRM_ID','name','email','birth','birth1','birth2','age']
f=open(r'C:\Users\wellslu\Desktop\CRM全客戶.csv',encoding='utf-8')
crm=pd.read_csv(f,chunksize=10000, error_bad_lines = False,names=head,low_memory=False)
```

```python
crm_n=crm.get_chunk(3058888)
```

```python
crm_n=crm_n.sort_values(by=['CRM_ID','age','birth','birth1','birth2']).reset_index(drop=True)
```

```python
crm_n=crm_n.drop_duplicates(['CRM_ID','name','email'], keep='first', inplace=False).reset_index(drop=True)
```

```python
crm_n.to_csv(r'C:\Users\wellslu\Desktop\crm_all.csv',index=False,encoding='utf-8')
```
