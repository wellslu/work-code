# CRM整理

```python
from tqdm import tqdm_notebook
import pandas as pd
```

```python
head=['CRM_ID','name','email','birth','birth1','birth2','age']
f=open(r'C:\Users\wells\OneDrive\Desktop\CRM全客戶.csv',encoding='utf-8')
crm=pd.read_csv(f, error_bad_lines = False,names=head)
```

```python
crm=crm.sort_values(by=['CRM_ID','age','birth','birth1','birth2']).reset_index(drop=True)#排序

crm=crm.drop_duplicates(['CRM_ID','name','email'], keep='first', inplace=False).reset_index(drop=True)
#刪除'CRM_ID','name','email'重複的值
```

```python
crm1=crm.drop(['email','birth','birth1','birth2'],axis=1)#刪掉欄位

crm1=crm1.dropna(axis=0,how='any').reset_index(drop=True)#刪除年齡為空值
```

```python
def chan_type(aa):
    try:
        bb =int(aa['age'])
        return bb
    except:
        return None
```

```python
crm['age'] = crm.apply(chan_type,1)

crm_r = crm[crm['age'].notnull()]
crm_e = crm[crm['age'].isnull()]

crm_r=crm_r.drop(['email','birth','birth1','birth2'],axis=1)
```

```python
crm_e = crm_e.drop(['birth'],axis=1)

crm_e1 = crm_e[pd.notnull(crm_e['birth1'])].reset_index(drop=True)
crm_e1 = crm_e1.drop(['birth2','age'],axis=1)

crm_e2 = crm_e[pd.notnull(crm_e['birth2'])].reset_index(drop=True)
crm_e2 = crm_e2.drop(['birth1','age'],axis=1)

def chan_type1(aa):
    try:
        bb =pd.to_datetime(aa['birth1'])
        return bb
    except:
        return None

crm_e1['birth1']=crm_e1.apply(chan_type1,1)

crm_e1[crm_e1['birth1'].isnull()]
```

```python
crm_e1 = crm_e1.drop([25798]).reset_index(drop=True)

crm_e2[crm_e2['birth2'].isnull()]

crm_e1['birth1'] = crm_e1['birth1'].dt.year

crm_e1['birth1'] = 2019-crm_e1['birth1']

crm_e1=crm_e1.rename(columns={'birth1':'age'})
```

```python
def chan_type2(aa):
    try:
        bb =pd.to_datetime(aa['birth2'])
        return bb
    except:
        return None

crm_e2['birth2']=crm_e2.apply(chan_type2,1)

crm_e2[crm_e2['birth2'].isnull()]
```

```python
crm_e2 = crm_e2.drop([4312,16024]).reset_index(drop=True)

crm_e2['birth2'] = crm_e2['birth2'].dt.year

crm_e2['birth2'] = 2019-crm_e2['birth2']

crm_e2=crm_e2.rename(columns={'birth2':'age'})
```

```python
crm_ea = pd.concat([crm_e1,crm_e2],ignore_index=True)

crm_ea = crm_ea.drop_duplicates(['CRM_ID','name','email'], keep='first', inplace=False).reset_index(drop=True)

crm_ea = crm_ea.drop(['email'],axis=1)

crm_final=pd.concat([crm_r,crm_ea],ignore_index=True)

def change(aa):
    try:
        bb =int(aa['age'])
        return bb
    except:
        return None

crm_final['age']=crm_final.apply(change,1)

crm_final= crm_final[crm_final['age'].isin([n for n in range(15,86)])].reset_index(drop=True)

crm_1585= crm_final[crm_final['age'].isin([n for n in range(15,45)]+[f for f in range(61,86)])].reset_index(drop=True)

crm_4560= crm_final[crm_final['age'].isin([n for n in range(45,61)])].reset_index(drop=True)

crm_1585id=list(set(crm_1585['CRM_ID']))

crm_4560id=list(set(crm_4560['CRM_ID']))
```

# 類別次數字典1(主頻道分類)

```python
cato_all=[]
for i in tqdm_notebook(cat1['masterChannelTitle']):
    cato_all.append(i)

cato_count=[]
cato_alls=list(set(cato_all))
for i in tqdm_notebook(cato_alls):
    cato_count.append(cato_all.count(i))

c_count1=dict(zip(cato_alls,cato_count))
```

```python
cato_all=[]
for i in tqdm_notebook(cat1['subChannelTitle']):
    cato_all.append(i)
cato_count=[]
cato_alls=list(set(cato_all))
for i in tqdm_notebook(cato_alls):
    cato_count.append(cato_all.count(i))

sub_count1=dict(zip(cato_alls,cato_count))
```

# 15到45歲+60到85歲主頻道分析

```python
def abc1(aa):
    caty=[]
    for i in tqdm_notebook(aa['pid']):
        try:
            caty.append(cat1[cat1['id'].isin([str(i)])].reset_index(drop=True)['masterChannelTitle'][0])
        except:
            1
    cat_count=[]
    caty_all=list(set(caty))
    for i in tqdm_notebook(caty_all):
        cat_count.append(caty.count(i))
    county=pd.DataFrame({'word':caty_all,
                        'count':cat_count})
    county=county.sort_values(by=['count'],ascending=False).reset_index(drop=True)
    aaa=[]
    for i in tqdm_notebook(range(len(county['count']))):
        aaa.append(county['count'][i]/c_count1[county['word'][i]])
    county['平均點擊']=aaa
    return county
```
